This notebook allows users to compare any 2 teams. First, several graphs compare both teams' stats, so the user can see how they match up. Then, the program simulates a playoff series between them. The function `n_simulations` runs many simulations (default = 25) of this playoff series and prints both teams' probabilities of winning it.

The user should modify `team` and `opp` in the second code cell, as well as `n` for the number of simulations they want to run.

```python
import numpy as np
import pandas as pd
import sys
import os
import matplotlib.pyplot as plt

sys.path.append("../src")

from fetch_gamelog import gamelog
from predict_result import simulate_series, n_simulations
```

```python
team = "" #user input
opp = "" #user input

team_gamelog = gamelog(team)
opp_gamelog = gamelog(opp)

def length(df):
    return df[("Score", "Rslt")].shape[0]
```

```python
#Points through the season

plt.figure(figsize = (10, 5))
plt.step(team_gamelog[("Unnamed: 2_level_0", "Date")], team_gamelog[('Unnamed: 5_level_0', 'Points')], where='post', label=team)
plt.step(opp_gamelog[("Unnamed: 2_level_0", "Date")], opp_gamelog[('Unnamed: 5_level_0', 'Points')], where='post', label=opp)

plt.title('Points Through Season')
plt.grid(True, axis='both', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
#Comparison of Shots on Goal

data1 = team_gamelog[('Team', 'SOG')]
data2 = opp_gamelog[('Team', 'SOG')]
plt.hist(data1, label = team, alpha = 0.5); 
plt.hist(data2, label = opp, alpha = 0.5);
plt.xlabel("Shots on Goal");
plt.ylabel("# of Games");
plt.legend();
plt.show();
```

```python
#Comparison of Power Play Goals vs. Opportunities
def pp_success(ppg, ppo):
    return ppg / ppo

data1 = pp_success(team_gamelog[('Team', 'PPG')], team_gamelog[('Team', 'PPO')])
data2 = pp_success(opp_gamelog[('Team', 'PPG')], opp_gamelog[('Team', 'PPO')])      
plt.hist(data1, label = team, alpha = 0.5); 
plt.hist(data2, label = opp, alpha = 0.5);
plt.xlabel("Power Play Goals vs. Opportunities");
plt.ylabel("# of Games");
plt.legend();
plt.show();
```

```python
#Comparison of Save %
def save_pct(ga, sog):
    return 1 - (ga / sog)

data1 = save_pct(team_gamelog[('Score', 'GA')], team_gamelog[('Opponent', 'SOG')])
data2 = save_pct(opp_gamelog[('Score', 'GA')], opp_gamelog[('Opponent', 'SOG')])
plt.hist(data1, label = team, alpha = 0.5); 
plt.hist(data2, label = opp, alpha = 0.5);
plt.xlabel("Save %");
plt.ylabel("# of Games");
plt.legend();
plt.show();
```
